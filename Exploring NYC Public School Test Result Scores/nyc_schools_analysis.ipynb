# NYC Schools SAT Analysis

```python
import pandas as pd
schools = pd.read_csv("schools.csv")
schools.head()
```

## Prepare Total SAT Score

```python
schools['total_SAT'] = (
    schools['average_math'] + schools['average_reading'] + schools['average_writing']
)
schools.isna().sum()
```

## Best Schools in Math (>= 80% of 800)

```python
threshold = 0.8 * 800
best_math_schools = (
    schools[schools['average_math'] >= threshold]
    .sort_values('average_math', ascending=False)[['school_name', 'average_math']]
)
best_math_schools
```

## Top 10 Schools by Total SAT Score

```python
top_10_schools = (
    schools[['school_name', 'total_SAT']]
    .sort_values('total_SAT', ascending=False)
    .head(10)
)
top_10_schools
```

## Borough With Largest Standard Deviation in Total SAT

```python
borough_stats = schools.groupby('borough')['total_SAT']
std_series = borough_stats.std()
max_std_borough = std_series.idxmax()
num_schools = schools[schools['borough'] == max_std_borough].shape[0]
average_SAT = borough_stats.mean()[max_std_borough]
std_SAT = std_series[max_std_borough]
largest_std_dev = pd.DataFrame({
    'borough': [max_std_borough],
    'num_schools': [num_schools],
    'average_SAT': [round(average_SAT, 2)],
    'std_SAT': [round(std_SAT, 2)]
})
largest_std_dev
```
