## 1️⃣ Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2️⃣ Load Dataset

```python
# Update the path if needed
crimes = pd.read_csv("../data/crimes.csv")

crimes.head()
```

## 3️⃣ Data Preparation

```python
# Extract hour if not already available
crimes['HOUR OCC'] = crimes['TIME OCC'].str[:2].astype(int)
crimes.info()
```

## 4️⃣ Peak Crime Hour

```python
peak_crime_hour = crimes['HOUR OCC'].mode()[0]
peak_crime_hour
```

```python
OCC = crimes['HOUR OCC'].value_counts().sort_index()
OCC_df = OCC.reset_index()
OCC_df.columns = ['HOUR OCC', 'count']
sns.lineplot(data=OCC_df, x='HOUR OCC', y='count')
plt.title('Number of Crimes by Hour of Occurrence')
plt.xlabel('Hour of Occurrence')
plt.ylabel('Number of Crimes')
plt.xticks(range(0, 24))
plt.grid()
plt.show()
```

## 5️⃣ Area with Most Night Crimes (10PM–3:59AM)

```python
night_crimes = crimes[
    (crimes['HOUR'] >= 22) | (crimes['HOUR'] <= 3)
]

peak_night_crime_location = (
    night_crimes['AREA NAME']
    .value_counts()
    .idxmax()
)

peak_night_crime_location
```

## 6️⃣ Victim Age Group Distribution

```python
bins = [0, 17, 25, 34, 44, 54, 64, 100]
labels = ["0-17", "18-25", "26-34", "35-44",
          "45-54", "55-64", "65+"]

crimes['AGE GROUP'] = pd.cut(
    crimes['Vict Age'],
    bins=bins,
    labels=labels
)

victim_ages = crimes['AGE GROUP'].value_counts().sort_index()
victim_ages
```

## 7️⃣ Conclusion

- Identified the peak crime hour.
- Determined which area experiences the highest number of night crimes.
- Analyzed victim age distribution using categorical binning.
